```python
import os
import sqlite3
con = sqlite3.connect('database.sqlite')
cursorObj = con.cursor()
cursorObj.execute("select strftime('%d/%m/%Y',date) as Data ,season as Season, name as League_Name, max(home_team_goal+away_team_goal) as Goal_scored From League, Match where Match.country_id = League.country_id group by season,name order by Match.country_id;")
rows = cursorObj.fetchall()

for row in rows:
  print(row)
```
